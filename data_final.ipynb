# נוסחאת מרחק

```python
def distance2(s_lat, s_lng, e_lat, e_lng):

   
   R = 6373.0

   s_lat = s_lat*np.pi/180.0                      
   s_lng = np.deg2rad(s_lng)     
   e_lat = np.deg2rad(e_lat)                       
   e_lng = np.deg2rad(e_lng)  

   d = np.sin((e_lat - s_lat)/2)**2 + np.cos(s_lat)*np.cos(e_lat) * np.sin((e_lng - s_lng)/2)**2

   return 2 * R * np.arcsin(np.sqrt(d))
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
chi_selector = SelectKBest(chi2, k=4)
chi_selector.fit(X, y)
chi_feature = chi_selector.get_feature_names_out()
print("The top 4 top scored features are ", chi_feature)
```

## Choosing a K Value
**13. Find the optimal number of neighboors**


```python
import numpy as np
error_rate = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != train_results['actual']))
```

```python
trace = go.Scatter(
    x=[i for i in range(1,100)],
    y=error_rate,
    mode='markers+lines', 
)
fig = go.Figure(trace)
fig.update_layout(
    title='Error by K Value', 
    xaxis_title='k neighboors',
    yaxis_title='Error rate'
)

fig.show()
```
